```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
import pandas as pd
```

```python
x=iris.data
```

```python
x.shape
```

```python
y=iris.target
```

```python
y.shape
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
model1=AgglomerativeClustering(n_clusters=3)
```

```python
model1.fit(x)
```

```python
lab_mod1=model1.labels_
```

```python
import matplotlib.pyplot as plt
```

```python
dendrogram(linkage(x,method='ward'))
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model2=KMeans(n_clusters=3)
```

```python
model2.fit(x)
```

```python
lab_mod2=model2.labels_
```

```python
dendrogram(linkage(x,method='ward'))
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
s1=silhouette_score(x,model1.fit_predict(x))
s2=silhouette_score(x,model2.fit_predict(x))
```

```python
s1,s2
```

```python
plt.bar(['Aggo','KMeans'],[s1,s2])
plt.xlabel('Models')
plt.ylabel('Comp b/w agg,kmeans')
plt.show()
```
